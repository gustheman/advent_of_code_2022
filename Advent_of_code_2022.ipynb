```python
wins = ["A Y", "B Z", "C X"]
lose = ["A Z", "B X", "C Y"]
# draw = ["A Z", "B X", "C Y"]
points = {
    "X": 1,
    "Y": 2,
    "Z": 3,
    "A": 1,
    "B": 2,
    "C": 3,
}

to_win = {
    "A": "Y",
    "B": "Z",
    "C": "X",
}

to_lose = {
    "A": "Z",
    "B": "X",
    "C": "Y",
}

to_draw = {
    "A": "X",
    "B": "Y",
    "C": "Z",
}

def calculate_points(play):

  mp = play[2]
  # print(play, mp)
  r = points[mp]
  if play in wins:
    r += 6
  elif play in lose:
    r += 0
  else:
    r += 3
  return r

def calculate_points_p2(play):

  mp = play[2]
  # print(play, mp)
  r = points[mp]
  if mp == "Z":
    play = f"{play[0]} {to_win[play[0]]}"
  elif mp == "X":
    play = f"{play[0]} {to_lose[play[0]]}"
  else:
    play = f"{play[0]} {to_draw[play[0]]}"

  return calculate_points(play)

lines = read_lines("/content/intput.txt")

total = 0
for l in lines:
  if l:
    c = calculate_points_p2(l.strip())
    total += c
    # print(l, c)

print(total)
```

```python
# lines = read_lines(input_path_test)
lines = read_lines(input_path)

foldersTotals = {}

class Folder():

  def __init__(self, name, parent):
    self.name = name
    self.parent = parent
    self.folders = {}
    self.files = {}

  def get_total_size(self):

    total = 0
    for k, v in self.files.items():
      total += v

    for k, v in self.folders.items():
      partial = v.get_total_size()
      total += partial

    foldersTotals[f"{self}"] = total
    return total

  def __str__(self):
    if not self.parent:
      return f"{self.name}"
    else:
      return f"{self.parent}{self.name}/"
    # print(f"{self.parent}/{self.name}")

dir_structure = {}


total_folders = 0
root = Folder("/", None)
current_path = []
current_folder = root
i = 0
while i < len(lines):
  line = lines[i]
  # print(line)

  if line.startswith("$ cd "):

    f = line.split()[-1]
    if f == "/":
      current_folder = root
    elif f == "..":
      current_folder = current_folder.parent
      # current_path = current_path[:-1]
    else:
      if not f in current_folder.folders.keys():
        print(f"found a new folder {f}")
        new_folder = Folder(f, current_folder)
      else:
        new_folder = current_folder.folders[f]
      current_folder = new_folder

    # print("found cd", line, current_folder)

  elif line.startswith("$ ls"):
    # print("found ls", line)
    i += 1
    continue
  else:
    tokens = line.split(" ")
    if tokens[0] == "dir":
      current_folder.folders[tokens[1]] = Folder(tokens[1], current_folder)
      total_folders += 1
    else:
      current_folder.files[tokens[1]] = int(tokens[0])
    # print("files in path:", current_folder, tokens[1], tokens[0])

  i += 1

print(root.folders)
print(root.files)
print(root.get_total_size())
print(foldersTotals)

values = [v for v in foldersTotals.values() if v <= 100000]

print(sum(values))
print(len(values), total_folders, len(foldersTotals))
```

##Problem 9: Rope Bridge

```python
import numpy as np

# lines = read_lines(input_path_test)
lines = read_lines(input_path)

def get_matrix_size(lines):
  r = 0
  mr = 0
  ml = 0
  u = 0
  mu = 0
  md = 0
  for l in lines:
    ts = l.split(" ")
    if ts[0] == "R":
      r += int(ts[1])
      if r > mr:
        mr = r
    if ts[0] == "L":
      r -= int(ts[1])
      if r < ml:
        ml = r
        # print("r is negative!!!")
    if ts[0] == "U":
      u -= int(ts[1])
      if u < md:
        md = u
    if ts[0] == "D":
      u += int(ts[1])
      if u > mu:
        mu = u

        # print("u is negative!!!")

    # print(ts)
  return mr, ml, mu, md


def simulate_path(lines, size, iniW, iniH):

  visited = np.full((size, size), False)
  matrix = np.full((size, size), '.')

  wT = iniW
  hT = iniH
  wH = iniW
  hH = iniH


  for l in lines:
    d, s = l.split(" ")
    s = int(s)
    for i in range(s):
      if d == "R":
        wH += 1
      if d == "L":
        wH -= 1
      if d == "U":
        hH -= 1
      if d == "D":
        hH += 1

      # move T
      # same row
      if (hH == hT) and (d in ["R", "L"]):
        if abs(wT - wH) > 1:
          if d == "R":
            wT += 1
          if d == "L":
            wT -= 1
      # same col
      elif (wH == wT) and (d in ["U", "D"]):
        if abs(hT - hH) > 1:
          if d == "U":
            hT -= 1
          if d == "D":
            hT += 1
      # diag
      else:

        dh = abs(hT - hH)
        dw = abs(wT - wH)
        # print('diagonal', dh, dw)
        if dh > dw:
          if hT > hH:
            hT = hH + 1
          else:
            hT = hH - 1
          wT = wH
        elif dh < dw:
          if wT > wH:
            wT = wH + 1
          else:
            wT = wH - 1
          hT = hH

      # print(wT, wH)
      visited[hT, wT] = True

    matrix[hH, wH] = 'H'
    # print(matrix)
    # print(visited)
    # print()

  num_visited = np.count_nonzero(visited)
  return num_visited


r, l, u, d = get_matrix_size(lines)
print(r, l, u, d)
iniW = (-l) + 1
iniH = (-d) + 1
side = max(iniW + r, iniH + u) + 1
print(side, iniW, iniH)
num_visited = simulate_path(lines, side, iniW, iniH)
print(num_visited)
# print(lines)
```
